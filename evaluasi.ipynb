```python
import seaborn as sns
from matplotlib import pyplot as plt
```

FD

```python
# tn, fp, fn, tp
cf1 = [[0,15], [2,16]]
cf2 = [[0,9], [3,8]]
cf3 = [[0,17], [3,10]]
cf4 = [[0,19], [2,8]]

fig, axes = plt.subplots(2, 2, figsize=(18, 10))

axes[0,0].set_title('FD1')
sns.heatmap(cf1, annot=True, ax=axes[0, 0]).set(xlabel="predicted", ylabel="actual")

axes[0,1].set_title('FD2')
sns.heatmap(cf2, annot=True, ax=axes[0, 1]).set(xlabel="predicted", ylabel="actual")

axes[1,0].set_title('FD3')
sns.heatmap(cf3, annot=True, ax=axes[1, 0]).set(xlabel="predicted", ylabel="actual")

axes[1,1].set_title('FD4')
sns.heatmap(cf4, annot=True, ax=axes[1, 1]).set(xlabel="predicted", ylabel="actual")
```
